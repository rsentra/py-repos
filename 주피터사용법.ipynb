# 패키지 생성 
* 패키지-모듈의 묶음  
   패키지에는 반드시  __init__.py가 있어야함. 
* import방법  
  from 패키지명 import 모듈명1, 모듈명2  
  from 패키지명 import *   
  '*' 로 모든 모듈을 import할때는 __init__.py에 __all__ 변수에  
  등록된 모듈만 대상임 (ex, __all__=['moduleA',moduleB'] )
  패키지명은 폴더명, 모듈은 .py로 된 파이썬 파일이어야 함
  
  * import os,sys
      sys.path - path확인  
      sys.path.append('C:/Users/user/python')  
      
 
 ## 파이썬 경로설정

### * 파이썬 경로가 다음과 같다고 가정  
   1.c:\python          - 파이썬 실행 메인 경로  
   2.c:\python\mypkgs   - 사용자 패키지 경로 및 mymod.py가 있을 경우   
     mymod.py  
```python  
        def common()  
          print('common modules')  
```  
   3.c:\python\myapps   - 사용자 파이썬 프로그램 경로  
   
### * 아래처럼 설정 후 사용
    1)python 과 mypkgs 경로 아래에 __init__py 생성 (파일 내용이 없어도 됨)  
    2)myapp 에 있는 파이썬 프로그램에서 import  
      [방법1]  
      from  mypkgs import myMod  
      mymod.common()  
      [방법2]    
      from  mypkgs.myMod import common  
      common()  
      [방법3]  
      import mypkgs.myMod as mod  
      mod.common()

# 주피터 실행경로변경  
1. 주피터 노트북 아이콘 우클릭  
2. 파일위치 열기  
3. 파일중 Jupyter NoteBook(anaconda3) 우클릭->속성  
4. 대상의 '%USERPROFILE%/' 부분을 다른경로(ex,'E:\') 로 변경  
   내PC -> 속성 -> 고급시스템설정 -> 환경변수 -> 시스템변수 -> PYTHONPATH를 'E:\python' 처럼 잡아줌  
5. 재실행  


이렇게해서 패키지가 열리지 않는 오류가 생기면  
1. anaconda prompt 열기  
2. cmd> jupyter notebook --generate-config
3. 위 에서 생기는 파일경로 및 파일복사  
   C:\Users\User\.jupyter\jupyter_notebook_config.py  
   
# 주피터 랩 사용  
1. 설치: pip install jupyterlab  
2. 실행: jupyter lab 
3. 바로가기 아이콘 작성  
  1) 주피터 노트북 아이콘 우클릭    
  2) 파일위치 열기   
  3) 주피터 노트북 바로가기를 복사하여 주피터 랩으로 변경  
  4) 속성 클릭하여 바로가기 부분의 jupyter-notebook-script.py를 jupyter-lab-script.py로 변경  
4. 작성된 바로가기를 윈도웨 등록하여 사용

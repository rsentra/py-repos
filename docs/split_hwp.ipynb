```python
''' 새로 파일을 생성하여 복사하고 다른 파일로 저장 '''
from pyhwpx import Hwp
from time import sleep


def split_hwp(**kwargs):
    ''' hwp문서를 분할하는 함수
    input_file: 입력파일명, output_file: 출력파일명
    cover_pages: 커버페이지 리스트, title_pages: 타이틀페이지 리스트(재생성하므로 copy 제외)
    cntnt_pages: 문서 페이지 리스트 '''
    
    my_params = {}
    for key, value in kwargs.items():
        my_params[key] = value

      #  원본문서 열기
    hwp = Hwp()
    hwp.open(my_params['input_file'])  #invisible로 하면 반복작업하다가 오류발생함

    hwp_copy = Hwp(visible=False, new=True)
    # hwp_copy.set_font(
    #         Height=1000,
    #         FaceName="함초롱바탕",
    #     )

    #페이지는 표지+목차 페이지를 모두 합하여 설정해야 함
    cover_pages = my_params['cover_pages'] #[1,2]       #커버
    title_pages = my_params['title_pages'] #[3,4,5,6,7] #타이틀
    cntn_pages = my_params['cntnt_pages'] #[8,9,10,12]  #컨텐츠
    incl_pages = cover_pages + cntn_pages 

    hwp.switch_to(0)  # 원본 문서에서
    k = 0
    for i in range(1, hwp.PageCount+1):  # 1~100페이지까지
        # sleep(0.5)
        if i in incl_pages:
            hwp.goto_page(i)  # i번째 페이지로 이동
            hwp.CopyPage()  # 페이지 복사해서
            hwp_copy.PastePage()  # 끼워넣기 한 후
            k += 1
            if k == 1:
                hwp_copy.MoveDocBegin()  # 기존의 1페이지(빈 페이지)로 가서
                hwp_copy.DeletePage()  # 빈 페이지 삭제 후

    hwp_copy.PageNumPos(side_char=True) #페이지반호 삽입
    p = max(cover_pages) + 1  #제목차례를 만들 위치
    hwp_copy.goto_page(p)
    hwp_copy.Run('BreakPage')  #페이지분할
    hwp_copy.goto_page(p)

    hwp_copy.save_as(my_params['output_file'] ) #(f"output/out02.hwp")  # 새 페이지 저장

    hwp.close()  # 변경내용 버리기
    # hwp.quit()   # 백그라운드의 한/글 프로그램 종료
    hwp_copy.close()
    hwp_copy.quit()
    print("split to...",incl_pages)
    return None


test_yn = False
if test_yn:
    path = "C:/Users/hkccloud01/Downloads/범정부 시스템 메뉴 역할/사용자지침서/"
    file = '2024GW-APP-5020-사용자지침서-3.AP(2.KMS)-v1.0.hwp'
    inf = path + file
    split_hwp(input_file=inf, cover_pages=[1,2],title_pages=[3,4,5,6,7],cntnt_pages=[8,9], output_file="output/out01.hwp" )
    split_hwp(input_file=inf, cover_pages=[1,2],title_pages=[3,4,5,6,7],cntnt_pages=[8,9], output_file="output/out02.hwp" )
    split_hwp(input_file=inf, cover_pages=[1,2],title_pages=[3,4,5,6,7],cntnt_pages=[8,9], output_file="output/out03.hwp" )
```

```python
''' 제목만들기 함수 '''
import pyautogui
from pyhwpx import Hwp
from time import sleep

def find_word(hwp, 단어):
    hwp.HAction.GetDefault("RepeatFind", hwp.HParameterSet.HFindReplace.HSet)
    hwp.HParameterSet.HFindReplace.FindString = 단어
    hwp.HParameterSet.HFindReplace.Direction = hwp.FindDir("Forward")
    hwp.HParameterSet.HFindReplace.IgnoreMessage = 1
    hwp.HParameterSet.HFindReplace.FindType = 1
    return hwp.HAction.Execute("RepeatFind", hwp.HParameterSet.HFindReplace.HSet)

def make_title(doc, mode=False):
    # print('start---')
    hwp = Hwp(visible=True) 
    hwp.open(doc) 

    res = find_word(hwp,"제목 차례")
    if res:
        print('title existed.... at',doc)
        hwp.close()
    else:
        key = 'F10' 
        pyautogui.typewrite([key])  #메뉴

        key = 'right'
        pyautogui.press(key, 19)  #도구메뉴까지 화살표

        #도구 메뉴내 단축키
        pyautogui.press('C') #차례/색인
        pyautogui.press('T') #차례만들기
        pyautogui.hotkey('alt', 't') #표 차례
        pyautogui.hotkey('alt','g') #그림 차례
        pyautogui.hotkey('alt','m') #수식차례
        pyautogui.press('d') #만들기 버튼
        # pyautogui.press('enter')
        sleep(5)   #제목생성을 위한 시간
        if mode:
            ss = hwp.clear(2)
            print('title Create & save---- at',doc, 'result= ', ss)
        else:
            ss = hwp.close()
            print('title Create & discard---- at',doc)
        # hwp.close()
        
    hwp.quit()
    return True
    
test_yn = False
if test_yn:
    make_title(f"output/out03.hwp", mode=True)
    make_title(f"output/out04.hwp", mode=True)
    make_title(f"output/out02.hwp", mode=True)
```

```python
import pyautogui
import time
waitTime=5
conf=0.7
pyautogui.PAUSE=0.3

#이미지읽기 - 반복
def f_readImage(imgStr,waitCnt):
    retry_counter = 0
    while retry_counter < waitCnt:
        try:
            result = pyautogui.locateOnScreen(imgStr)
            # result = pyautogui.locateOnScreen(imgStr,confidence=conf)
            if result:
                time.sleep(1)
                print('successful read = ',retry_counter, '- ', imgStr)
                retry_counter = 10  # to break the loop
                pyautogui.click(result)
                return result
            else:
                retry_counter += 1
        except:
            time.sleep(1)  # retry after some time, i.e. 1 sec
            retry_counter += 1
    print('fail to read image ', retry_counter, '- ', imgStr)
    # logging.info(f'fail to read image {retry_counter} - {imgStr}')
    return None
```

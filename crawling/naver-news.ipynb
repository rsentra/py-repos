```python
import urllib.request as req
import bs4

url='https://news.naver.com'
html=req.urlopen(url)
bsObj=bs4.BeautifulSoup(html.read(),'html.parser')


ul=bsObj.find("ul",{"class":"hdline_article_list"})

lis=ul.findAll('li')
titles=[li.find("a").text for li in lis]
for title in titles:
     print(title.strip())
```

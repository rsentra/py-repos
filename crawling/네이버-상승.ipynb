```python
#네이버 상승종목
import requests as req
import bs4 

def getCrawl(url):
    data=req.get(url)
    return data.content

def getParse(pageString):
    bs_Obj=bs4.BeautifulSoup(pageString,'html.parser')
    box_type_1=bs_Obj.find('div',{'class':'box_type_l'})
    type_2=box_type_1.find('table',{'class':'type_2'})
   # print(type_2)
   # tbody=type_2.find('tbody')
  #  print(tbody)
    trs=type_2.findAll("tr")
   # print(trs)
    stockInfos=[]
    for tr in trs:
        try:
            stockInfo=getInfo(tr)
            #print(stockInfo)
            stockInfos.append(stockInfo)
        except Exception as e:
            pass
    print(stockInfos)
    return stockInfos

def getInfo(tr):
    tds=tr.findAll('td')
    rank=tds[0].text
    aTag=tds[1].find('a')
    href=aTag["href"]
    name=aTag.text
    nowPrice=tds[2].text
    rate=tds[4].text.replace('\n','')
    rate=rate.replace('\t','')
    volume=tds[5].text
    return {"rank":rank,"name":name,"href":href,"code":href[-6:],"nowprice": nowPrice,"rate": rate, "volume":volume}

def getSiseRise(sosok):
    url="https://finance.naver.com/sise/sise_rise.nhn?sosok={}".format(sosok)
    pageString=getCrawl(url)
    list = getParse(pageString)
    return list


#resultR=[]
ResultR=getSiseRise(0)

print(ResultR)

print('2수집개수는=', len(ResultR))
import pandas as pd

df2=pd.DataFrame(ResultR)
```

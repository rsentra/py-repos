## 여러 페이지 저장- tif만  

```python
from PIL import ImageSequence, Image
import glob
import sys
import os
indir='e:/python/5.Exercise on Glad'

def create_aggregated_tiff(input_tiffs):
    im_list = []
    for tiff_path in input_tiffs:
        print(tiff_path)
        img = Image.open(open(tiff_path,'rb'))
        for page in ImageSequence.Iterator(img):
            im_list.append(page)

    print(im_list[0])
    return im_list
    #im_list.save("test.tif", compression="tiff_deflate", save_all=True)

if __name__ == '__main__':
    input_tiffs = []
#     input_tiffs = ['./ff1.png',
#                    './ff2.png',
#                    './ff3.png',
#                    './ff4.png'
#                   ]
    for fname in glob.glob(os.path.join(indir, 'ff*.png')):
#         print(fname)
        input_tiffs.append(fname)
    
    lst = create_aggregated_tiff(input_tiffs=input_tiffs)
    img1 = lst[0]
    img1.save("reportAll.tiff", save_all=True, append_images=lst[1:])
```

```python
##여러장 저장 안됨
import sys #매개변수를 입력받기 위해 사용
from PIL import Image

def ClearWhiteBackground(image):
    #RGBA로 속성 변경
    image = image.convert('RGBA')

    #해당 이미지의 배열 받아오기
    imageData = image.getdata()

    newImageData = []

    for pixel in imageData:
        if pixel[0] > 240 and pixel[1] > 240 and pixel[2] > 240:
            # rgb값이 240,240,240 이상일 경우(흰색에 가까울 경우) 알파값을 0으로 준다
            newImageData.append((0,0,0,0))
        else:
            # 아닐경우 그대로 쓴다
            newImageData.append(pixel)

    #이미지를 덮어 씌움
    image.putdata(newImageData)

    return image

from reportlab.pdfgen.canvas import Canvas

def ImageToPDF(imagePath, pdfPath):
    newCanvas = Canvas(pdfPath, pagesize=Image.open(imagePath,'r').size)

    newCanvas.drawImage(image=imagePath,x=0,y=0, mask='auto')

    newCanvas.save()

def Test():
#     argList = sys.argv
#     dirn=argList[1]
#     filen=argList[2]

    '''
    매개변수에 두번째는 이미지가 있는 경로, 세번째는 이미지의 이름이 들어온다
    첫번째는 파일의 경로 고정
    ex) 'C:\\ImageToPDF.py' 'C:\\ImagePath\\' 'sample.jpg'
    '''

    dirn='e:/python/5.Exercise on Glad/'
    filen='reportAll.tif'
       
    image = Image.open(dirn+filen,'r')
    clearImage = ClearWhiteBackground(image)

    clearImage.save(dirn+'clearSample.png')

    ImageToPDF(dirn+'clearSample.png', dirn+'watermarkImage.pdf')

Test()
```

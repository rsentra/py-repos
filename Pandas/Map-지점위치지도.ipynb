```python
import pandas as pd
#df_my=pd.DataFrame({'a':['b1','b2','b3','b4','b5'],
#                    '위도':[37.62048886,37.61322334,37.613223355,37.61324335,37.61322345], 
#                 '경도':[126.9342435,126.9195884,126.9495884,126.9595884,126.9505884]})

df_my=pd.read_excel('../testData/지점주소.xlsx')
```

```python
geo_df=df_my

map = folium.Map(location=[geo_df['위도'].mean(),geo_df['경도'].mean()+0.25], zoom_start=10)
             #    tiles='Stamen toner',fill_color = 'YlGnBu')

for n in geo_df.index:
    popup_name = geo_df.loc[n, 'brh']+'-'+geo_df.loc[n, 'address']
    if geo_df['gb'][n] == '지사' :
        icon_color = 'darkgreen'
    else:
        icon_color = 'blue'    
    lat=geo_df.loc[n, '위도']
    long=geo_df.loc[n, '경도']
    size=int(geo_df.loc[n, 'size'])
    folium.CircleMarker(
        location=[lat, long],
        radius=size,    
       # popup=popup_name,show=True,
        color= icon_color,
        tooltip=popup_name,stickey=True,
        draggable=True,
        fill=True,
        fill_color=icon_color
    ).add_to(map)

map.save('글래드지점Map.html')
map
```
